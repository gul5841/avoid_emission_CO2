```python
import pandas as pd

# Dosya yolu
file_path = r"C:\Users\w11\dergi2\processed_data\final\12_final.xlsx"

# Veriyi oku
df = pd.read_excel(file_path)

# Sütun isimlerindeki olası gizli boşlukları temizleyelim (KeyError hatasını önlemek için)
df.columns = df.columns.str.strip()

# İstenen formatta dikey istatistiksel tablo oluşturan fonksiyon
def generate_statistical_table(dataframe, column_name):
    # Sütunun var olup olmadığını kontrol et
    if column_name not in dataframe.columns:
        return f"Hata: '{column_name}' sütunu Excel dosyasında bulunamadı. Lütfen ismini kontrol edin."
        
    # Sütundaki verileri tanımla
    desc = dataframe[column_name].describe()
    
    # İstenen dikey tablo yapısı
    table_data = {
        "Statistic": ["Data", "Mean", "Standard Deviation", "Minimum", 
                      "25th Percentile", "50th Percentile", "75th Percentile", "Maximum"],
        "Value": [
            desc['count'], 
            desc['mean'], 
            desc['std'], 
            desc['min'], 
            desc['25%'], 
            desc['50%'], 
            desc['75%'], 
            desc['max']
        ]
    }
    
    result_df = pd.DataFrame(table_data)
    return result_df

# Analiz edilecek sütun (Dosyanızda kesin olan bir sütun ismini buraya yazın)
# Örnek olarak 'generation_Solar' yazdım, kendi isteğinize göre değiştirin.
column_to_analyze = 'generation_Solar' 

# Fonksiyonu çalıştır
table_2 = generate_statistical_table(df, column_to_analyze)

# Eğer dönen sonuç bir DataFrame ise (hata mesajı değilse) yazdır
if isinstance(table_2, pd.DataFrame):
    # Format ayarı (binlik ayracı olarak virgül ve 2 ondalık basamak)
    pd.options.display.float_format = '{:,.2f}'.format
    
    # İstenilen tam görünüm için yazdır (index=False ile baştaki 0,1,2.. sayılarını gizleriz)
    print("Table 2. Statistical Table")
    print(table_2.to_string(index=False))
else:
    print(table_2) # Hata mesajını yazdırır
```
