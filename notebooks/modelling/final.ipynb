```python
import pandas as pd
import os

# =====================================================
# DOSYA YOLLARI
# =====================================================

base_dir = r"C:\Users\ismailgulsoy\dergi2"

capacity_file = os.path.join(
    base_dir,
    "processed_data",
    "steps",
    "09_capacity_generation_population.xlsx"
)

weather_file = os.path.join(
    base_dir,
    "processed_data",
    "steps",
    "10_sun_rain_wind.xlsx"
)

output_file = os.path.join(
    base_dir,
    "processed_data",
    "final",
    "final_dataset.xlsx"
)

# =====================================================
# DOSYALARI OKU
# =====================================================

df_capacity = pd.read_excel(capacity_file)
df_weather = pd.read_excel(weather_file)

# =====================================================
# SÜTUN İSİMLERİNİ TEMİZLE
# =====================================================

df_capacity.columns = df_capacity.columns.str.strip()
df_weather.columns = df_weather.columns.str.strip()

# =====================================================
# BİRLEŞTİR
# =====================================================

df_final = pd.merge(
    df_capacity,
    df_weather,
    on=["Province", "Year", "Month"],
    how="left"
)

# =====================================================
# SIRALA
# =====================================================

df_final = df_final.sort_values(
    ["Plate", "Year", "Month", "Source"]
).reset_index(drop=True)

# =====================================================
# KAYDET
# =====================================================

df_final.to_excel(output_file, index=False)

# =====================================================
# RAPOR
# =====================================================

print("Tamamlandı.")
print("Satır Sayısı:", len(df_final))
print("Sütun Sayısı:", len(df_final.columns))
print("Dosya:", output_file)

print("\nEksik iklim verileri:")
print(
    df_final[
        ["Rainfall", "Solar_Radiation", "Temperature", "Wind_Speed"]
    ].isna().sum()
)

print("\nİlk 5 Satır:")
print(df_final.head())
```
