```python
import pandas as pd
import requests
import time

# =====================================================
# DOSYA YOLU
# =====================================================

coord_file = r"C:\Users\ismailgulsoy\dergi2\processed_data\steps\09_geo_metadata_cleaned.xlsx"

out_solar = r"C:\Users\ismailgulsoy\dergi2\raw_data\solar\Solar_Radiation.xlsx"
out_wind = r"C:\Users\ismailgulsoy\dergi2\raw_data\wind\Wind_Speed.xlsx"
out_rain = r"C:\Users\ismailgulsoy\dergi2\raw_data\rain\Rainfall.xlsx"
out_temp = r"C:\Users\ismailgulsoy\dergi2\raw_data\temperatur\Temperature.xlsx"

# =====================================================
# KOORDINATLARI OKU
# =====================================================

coords = pd.read_excel(coord_file)

for col in ["Latitude", "Longitude"]:
    coords[col] = (
        coords[col].astype(str)
        .str.replace(",", ".", regex=False)
    )
    coords[col] = pd.to_numeric(coords[col], errors="coerce")

coords = coords.dropna(subset=["Latitude", "Longitude"])

# =====================================================
# LISTELER
# =====================================================

solar_list = []
wind_list = []
rain_list = []
temp_list = []

# =====================================================
# NASA API
# =====================================================

for _, row in coords.iterrows():

    province = row["Province"]
    lat = float(row["Latitude"])
    lon = float(row["Longitude"])

    print(f"{province} işleniyor...")

    url = (
        "https://power.larc.nasa.gov/api/temporal/monthly/point"
        "?parameters=ALLSKY_SFC_SW_DWN,WS10M,PRECTOTCORR,T2M"
        "&community=RE"
        f"&longitude={lon}"
        f"&latitude={lat}"
        "&start=2021"
        "&end=2025"
        "&format=JSON"
    )

    try:
        response = requests.get(url, timeout=60)

        if response.status_code != 200:
            print(f"Hata: {province}")
            continue

        data = response.json()["properties"]["parameter"]

        solar = data["ALLSKY_SFC_SW_DWN"]
        wind = data["WS10M"]
        rain = data["PRECTOTCORR"]
        temp = data["T2M"]

        for ym in solar.keys():

            ym = str(ym)
            year = int(ym[:4])
            month = int(ym[4:6])

            base = {
                "Province": province,
                "Year": year,
                "Month": month
            }

            solar_list.append({**base, "Solar_Radiation": solar[ym]})
            wind_list.append({**base, "Wind_Speed": wind[ym]})
            rain_list.append({**base, "Rainfall": rain[ym]})
            temp_list.append({**base, "Temperature": temp[ym]})

        time.sleep(0.3)

    except Exception as e:
        print(f"Hata: {province} -> {e}")

# =====================================================
# DATAFRAME
# =====================================================

df_solar = pd.DataFrame(solar_list)
df_wind = pd.DataFrame(wind_list)
df_rain = pd.DataFrame(rain_list)
df_temp = pd.DataFrame(temp_list)

# =====================================================
# 4 AYRI EXCEL DOSYASI
# =====================================================

df_solar.to_excel(out_solar, index=False)
df_wind.to_excel(out_wind, index=False)
df_rain.to_excel(out_rain, index=False)
df_temp.to_excel(out_temp, index=False)

print("\nBitti!")
print("Solar:", out_solar)
print("Wind:", out_wind)
print("Rain:", out_rain)
print("Temp:", out_temp)
```
