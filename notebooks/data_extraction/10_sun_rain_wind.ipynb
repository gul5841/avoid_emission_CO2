```python
"""
PIPELINE STAGE: Meteorological Data Integration, Multi-Source Fusion & Spatio-Temporal Harmonization

INPUT:
    raw_data/rain/Rainfall.xlsx
    raw_data/solar/Solar_Radiation.xlsx
    raw_data/temperatur/Temperature.xlsx
    raw_data/wind/Wind_Speed.xlsx

OUTPUT:
    processed_data/steps/10_Sun_Rain_Wind.xlsx

LIBRARIES: pandas, os

1. OBJECTIVE:
    Integrate multiple meteorological datasets (rainfall, solar radiation, temperature,
    and wind speed) into a unified spatio-temporal dataset. The pipeline aligns all
    variables on common geographic and temporal keys, resolves naming inconsistencies,
    and produces a clean environmental feature set for energy and climate analysis.

2. DATA INGESTION:
    A. Input Sources:
       - Rainfall dataset
       - Solar radiation dataset
       - Temperature dataset
       - Wind speed dataset

    B. Loading Strategy:
       Each dataset is loaded independently using pandas.read_excel().

3. DATA INTEGRATION (MULTI-WAY MERGING):
    A. Primary Join Keys:

           ["Province", "Year", "Month"]

    B. Merge Strategy:
       Stepwise outer joins:

           1. Rainfall ⟷ Solar Radiation
           2. Result ⟷ Temperature
           3. Result ⟷ Wind Speed

    C. Join Type:
       outer join is used to preserve all observations across datasets,
       ensuring no loss of meteorological records.

4. SPATIO-TEMPORAL ALIGNMENT:
    A. Alignment Dimensions:
       - Spatial: Province
       - Temporal: Year, Month

    B. Purpose:
       Ensure consistent cross-variable comparison across climate indicators.

5. DATA CLEANING & STANDARDIZATION:
    A. Province Normalization:
       Apply correction for inconsistent naming:

           "MERSİN(İÇEL)" → "MERSİN"

    B. Reason:
       Prevent duplication of geographic entities in downstream analysis.

6. DATA SORTING:
    A. Sorting Logic:
       Sort dataset by:

           Province → Year → Month

    B. Reset Index:
       Ensure sequential indexing after sorting.

7. DATA QUALITY HANDLING:
    A. Missing Values:
       Outer join may introduce NaN values; these are retained for completeness
       and future imputation stages.

8. OUTPUT DIRECTORY MANAGEMENT:
    A. Safety Check:
       Automatically create output directory if it does not exist:

           os.makedirs(..., exist_ok=True)

9. OUTPUT GENERATION:
    A. Final File:

           processed_data/steps/10_Sun_Rain_Wind.xlsx

    B. Export Format:
       - Excel (.xlsx)
       - No index column

10. PROCESS LOGGING & MONITORING:
    A. Runtime Outputs:
       - Total row count
       - Column count
       - File path confirmation
       - First 5 rows preview

11. EXPECTED OUTPUT DATASET:
    The resulting dataset includes:

        - Province-level meteorological indicators
        - Yearly and monthly temporal resolution
        - Rainfall data
        - Solar radiation data
        - Temperature data
        - Wind speed data

    This dataset is suitable for:

        - Renewable energy potential modeling
        - Climate-energy interaction studies
        - Machine learning forecasting
        - Environmental feature engineering
        - Spatio-temporal statistical analysis

"""


import pandas as pd
import os

# =====================================================
# DOSYA YOLLARI
# =====================================================

base_dir = r"C:\Users\w11\dergi2"

rain_path = os.path.join(
    base_dir,
    "raw_data",
    "rain",
    "Rainfall.xlsx"
)

solar_path = os.path.join(
    base_dir,
    "raw_data",
    "solar",
    "Solar_Radiation.xlsx"
)

temperature_path = os.path.join(
    base_dir,
    "raw_data",
    "temperatur",
    "Temperature.xlsx"
)

wind_path = os.path.join(
    base_dir,
    "raw_data",
    "wind",
    "Wind_Speed.xlsx"
)

final_path = os.path.join(
    base_dir,
    "processed_data",
    "steps","10_sun_rain_wind.xlsx"
)

# =====================================================
# DOSYALARI OKU
# =====================================================

df_rain = pd.read_excel(rain_path)
df_solar = pd.read_excel(solar_path)
df_temp = pd.read_excel(temperature_path)
df_wind = pd.read_excel(wind_path)

# =====================================================
# BİRLEŞTİR
# =====================================================

merge_keys = ["Province", "Year", "Month"]

df = df_rain.merge(
    df_solar,
    on=merge_keys,
    how="outer"
)

df = df.merge(
    df_temp,
    on=merge_keys,
    how="outer"
)

df = df.merge(
    df_wind,
    on=merge_keys,
    how="outer"
)

# =====================================================
# SIRALA
# =====================================================

df = df.sort_values(
    ["Province", "Year", "Month"]
).reset_index(drop=True)

# =====================================================
# KAYDET
# =====================================================


os.makedirs(os.path.dirname(final_path), exist_ok=True)

df["Province"] = df["Province"].replace({
    "MERSİN(İÇEL)": "MERSİN"})

df.to_excel(final_path, index=False)

print("Tamamlandı.")
print("Satır Sayısı:", len(df))
print("Sütun Sayısı:", len(df.columns))
print("Dosya:", final_path)

print("\nİlk 5 Satır:")
print(df.head())
```
