```python
"""
PIPELINE STAGE: Full Energy–Climate–Population Data Integration (Spatio-Temporal Feature Fusion)

INPUT:
    processed_data/steps/09_Capacity_Generation_Population.xlsx
    processed_data/steps/10_Sun_Rain_Wind.xlsx

OUTPUT:
    processed_data/final/11_Cap_Gen_Sun_Rain_Wind.xlsx

LIBRARIES: pandas

1. OBJECTIVE:
    Construct a unified spatio-temporal dataset by merging energy system data
    (capacity, generation, population) with meteorological variables (rainfall,
    solar radiation, temperature, wind speed). The resulting dataset enables
    integrated analysis of energy production, climate conditions, and demographic
    pressure at provincial level.

2. DATA INGESTION:
    A. Energy–Population Dataset (df1):
       Contains:
           - Province-level energy capacity (multiple sources)
           - Electricity generation (multiple sources)
           - Population data
           - Temporal identifiers (Year, Month)
           - Spatial identifier (Province)

    B. Meteorological Dataset (df2):
       Contains:
           - Rainfall
           - Solar Radiation
           - Temperature
           - Wind Speed
           - Province, Year, Month keys

3. DATA PREPROCESSING:
    A. String Normalization:
       Ensure consistency in spatial keys by applying:

           Province = str.strip()

       for both datasets.

    B. Feature Selection (Meteorological Data):
       Retain only relevant environmental variables:

           Province
           Year
           Month
           Rainfall
           Solar_Radiation
           Temperature
           Wind_Speed

4. DATA INTEGRATION (MERGE OPERATION):
    A. Join Strategy:
       Perform a left join:

           df1 LEFT JOIN df2

       using keys:

           ["Province", "Year", "Month"]

    B. Rationale:
       Preserve complete energy–population records while enriching them
       with available meteorological conditions.

5. SPATIO-TEMPORAL ALIGNMENT:
    A. Alignment Dimensions:
       - Spatial: Province
       - Temporal: Year, Month

    B. Objective:
       Ensure one-to-one correspondence across datasets for consistent panel structure.

6. DATA COMPLETENESS STRATEGY:
    A. Missing Weather Data:
       If meteorological observations are missing for certain records,
       they are retained as NaN values for downstream imputation or modeling.

7. OUTPUT GENERATION:
    A. Final Dataset:

           processed_data/steps/11_Cap_Gen_Sun_Rain_Wind.xlsx

    B. Format:
       - Excel (.xlsx)
       - No index column

8. PROCESS LOGGING:
    A. Runtime Outputs:
       - Final row count
       - Output file path confirmation

9. EXPECTED OUTPUT DATASET:
    The final dataset contains:

        - Energy system variables:
            * Installed capacity (by source)
            * Electricity generation (by source)

        - Socio-economic variable:
            * Population

        - Meteorological variables:
            * Rainfall
            * Solar radiation
            * Temperature
            * Wind speed

        - Spatio-temporal identifiers:
            * Province
            * Year
            * Month

    This dataset is designed for:

        - Renewable energy forecasting models
        - Climate-energy interaction analysis
        - Machine learning regression and classification
        - Spatio-temporal panel data modeling
        - Sustainability and energy transition studies

"""


import pandas as pd

# Dosyalar
file1 = r"C:\Users\w11\dergi2\processed_data\steps\09_capacity_generation_population.xlsx"
file2 = r"C:\Users\w11\dergi2\processed_data\steps\10_sun_rain_wind.xlsx"

output_file = r"C:\Users\w11\dergi2\processed_data\final\11_cap_gen_sun_rain_wind.xlsx"

# Oku
df1 = pd.read_excel(file1)
df2 = pd.read_excel(file2)

# İl isimlerini temizle
df1["Province"] = df1["Province"].astype(str).str.strip()
df2["Province"] = df2["Province"].astype(str).str.strip()

# file2'den alınacak sütunlar
weather_cols = [
    "Province",
    "Year",
    "Month",
    "Rainfall",
    "Solar_Radiation",
    "Temperature",
    "Wind_Speed"
]

df2 = df2[weather_cols]

# file1 sabit kalacak şekilde birleştir
df_final = pd.merge(
    df1,
    df2,
    on=["Province", "Year", "Month"],
    how="left"
)

# Kaydet
df_final.to_excel(output_file, index=False)

print(f"Tamamlandı. Satır sayısı: {len(df_final)}")
print(f"Dosya: {output_file}")
```
